```
from numpy.random import *
import numpy as np
```

```
class indian_poker():
  
  def __init__(self, num, sd1): #Distribute one playing card to each player.
    self.history=[]
    self.cards=list(range(1, 14))*4
    seed(sd1)
    self.num=num
    assert self.num<30, "Fuck!!! Too many players!!!" #The number of players must not be over 30. 
    self.hands=np.random.choice(self.cards, num)
    self.cards=[i for i in self.cards if not i in self.hands]
    self.your_num=randint(0, num, 1)[0]
    self.yours=self.hands[self.your_num]
    self.history.append(self.yours)

  def observe(self): #check other players' card
    print([self.hands[i] for i in range(len(self.hands)) if i!=self.your_num])
  
  def change(self, sd2=1): #Throw away my card and randomly choice a card from deck.
    seed(sd2)
    a=np.random.choice(self.cards, 1)[0]
    self.hands[self.your_num]=a
    self.cards.remove(a)
    self.yours=a
    self.history.append(a)
  
  def disclose(self): #disclose the result of this game
    a=sorted(self.hands, reverse=True)
    rank=a.index(self.yours)+1

    print("your rank is {}/{}".format(rank, self.num))
    print("your history is ", self.history)


    print("your card is {}".format(self.yours))
    print("All players' cards are", self.hands) 
```

```
game1=indian_poker(5, 6)
game1.observe()
```
